8.Understanding TODAY() and NOW()

a)What is the difference between TODAY() and NOW() in Excel? Provide an example of when you would use
 each function.

ans:

TODAY()

What it does:
TODAY() returns the current date only (no time).

When to use it:

Calculating age

Finding due dates

Tracking deadlines

Calculating number of days left

EXAMPLE:

=TODAY()

If A1 contains a due date (25-02-2026), you can calculate days left:
=A1-TODAY()

This tells you how many days are remaninig.


NOW()

What it does:
NOW() returns the current date AND current time.

Example Output:
20-02-2026 10:45 AM

When to use it:

->Time stamping entries

->Tracking login/logout time

->Measuring time differences

->Creating real-time dashboard
