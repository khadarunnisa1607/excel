6.Use a Slicer to filter data by Product and observe changes in sales.


ANS:
Screenshot 2026-02-24 204441.pngTo filter data by Product using a Slicer and observe changes in sales, follow these steps:

First, create a PivotTable using the dataset. Drag the Date field into Rows and group it by Months. Drag the Sales field into the Values area to calculate the total monthly sales.

Next, click anywhere inside the PivotTable and go to the PivotTable Analyze tab. Click on Insert Slicer and select the Product field. Click OK. A slicer will appear displaying the different products such as Mangoes, Oranges, and Apples.

When a product is selected in the slicer, the PivotTable and PivotChart automatically update to show sales data only for the selected product. This helps in comparing the sales performance of different products across months.

For example:

Selecting Mangoes displays only mango sales.

Selecting Oranges shows only orange sales.

Selecting Apples updates the sales accordingly.

Thus, a Slicer provides an interactive and visual way to filter PivotTable data and analyze changes in sales easily.

7.Why are Dashboards important for decision-making in business analysis?

ans:
Dashboards are important because they present large amounts of data in a simple, visual, and organized format. They help managers and analysts quickly understand business performance and make informed decisions.

Importance of Dashboards:

Provide real-time data visualization

Help identify trends and patterns

Support quick decision-making

Compare performance across regions, products, or months

Improve strategic planning

For example, a sales dashboard can show which region performs best, which product generates more revenue, and whether sales are increasing or decreasing monthly.

Thus, dashboards improve efficiency and accuracy in business analysis.
